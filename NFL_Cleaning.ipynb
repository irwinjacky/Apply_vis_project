```python
def read_NFL_data_in(teams):
  years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
  team_rosters = {}
  team_stats = {}
  team_odds = {}
  team_scores = {}
  team_injuries = {}

  for team in teams:
    roster_file_path = f'/content/drive/MyDrive/Data_vis_proj/NFLML/{team}/{team} Roster All Years.csv'
    team_rosters[team] = pd.read_csv(roster_file_path)
    team_rosters[team]['Team'] = team

    stat_path = f'/content/drive/MyDrive/Data_vis_proj/NFLML/{team}/{team} Stats by Year (PFR).csv'
    team_stats[team] = pd.read_csv(stat_path)
    team_stats[team]['Team'] = team

    odds_path = f'/content/drive/MyDrive/Data_vis_proj/NFLML/{team}/{team} WeeklyOdds All Years.csv'
    team_odds[team] = pd.read_csv(odds_path)
    team_odds[team]['Team'] = team

    scores_path = f'/content/drive/MyDrive/Data_vis_proj/NFLML/{team}/{team} WeeklyScores All Years.csv'
    team_scores[team] = pd.read_csv(scores_path)
    team_scores[team]['Team'] = team

    t_injuries = []

    for year in years:
      file_path = f'/content/drive/MyDrive/Data_vis_proj/NFLML/{team}/WeeklyInjuries/{team} {year} WeeklyInjuries.csv'
      injury_df = pd.read_csv(file_path)
      injury_df.columns = [col if col == 'Player' else f'{col} {year}' for col in injury_df.columns]
      t_injuries.append(injury_df)

    injuries = pd.merge(t_injuries[0], t_injuries[1], on='Player', how='outer')
    for i in range(2, (len(t_injuries))):
        injuries = pd.merge(injuries, t_injuries[i], on='Player', how='outer')

    team_injuries[team] = injuries

  return team_rosters, team_stats, team_odds, team_scores, team_injuries

```

```python
teams = set(weather['team_home'].unique())
teams2 = set(weather['team_away'].unique())
NFL_teams = list(teams.union(teams2))
print(NFL_teams)
```

```python
roster, stats, odds, scores, injuries = read_NFL_data_in(NFL_teams)
```

```python
for team in stats:
  stats[team].columns = ['Year' if col == 'Unnamed: 0' else col for col in stats[team].columns]

stats['RAM']
```

```python
imp_roster_columns = ['Player', 'Age', 'Pos', 'G', 'GS',
       'Yrs', 'AV', 'Year', 'Team']
opt_roster_columns = ['Wt', 'Ht']

imp_stats_columns = ['Points For', 'Points Against', 'Expected Wins', 'Expected Losses', 'Team']
opt_stats_columns = ['Division Placement', 'Points for Per Game', 'Points For Ranking',
       'Points Against Per Game', 'Points Against Ranking',
       'Team Rating SRS', 'SRS rank', 'Strength of Schedule']
imp_odds_columns = ['G#', 'Spread', 'Result', 'Year', 'Team']
opt_odds_columns = ['Opp']
imp_scores_columns = ['Date', 'Outcome', 'Score_Tm', 'Score_Opp', 'Year', 'Opp', 'Team']
opt_scores_columns = ['Week', 'Offense_TotYd', 'Defense_TotYd',
       'Expected Points_Offense', 'Expected Points_Defense',
       'Expected Points_Sp. Tms', 'Offense_TO', 'Defense_TO']
```

```python
def filter(info_dic, teams, cols, cols2):
  team_info = {}
  for team in teams:
    team_df = info_dic[team]
    team_df = team_df[(team_df['Year'] >= 2009) & (team_df['Year'] < 2021)]
    imp_team_df = team_df[cols]
    opt_team_df = team_df[cols2 + cols]

    team_info[team] = imp_team_df, opt_team_df
  return team_info
```

```python
filt_roster = filter(roster, NFL_teams, imp_roster_columns, opt_roster_columns)
filt_stats = filter(stats, NFL_teams, imp_stats_columns, opt_stats_columns)
filt_odds = filter(odds, NFL_teams, imp_odds_columns, opt_odds_columns)
filt_scores = filter(scores, NFL_teams, imp_scores_columns, opt_scores_columns)
```

```python
def injury_cleaning(info_dic, teams):
  team_info = {}
  for team in teams:
    team_df = info_dic[team]
    # Do we need to change these to match the Key from the nfl pfr website? Probable 75% chance of playing, Questionable 50% chance of playing, Doubtful 25% chance of playing, Out player is out,
    #    PUP player physically unable to perform list, IR player on the injuered reserve list.
    rep_values = {np.NaN: 0, 'S':1, 'Q':2, 'P':3, 'D':4,'O':5,'IR':6,'PUP':7, 'C19': 5, '/': 0, '//': 0, '//;90': 5}
    team_df = team_df.replace(rep_values)
    team_info[team] = team_df
  return team_info
```

```python
clean_inj = injury_cleaning(injuries, NFL_teams)
```

```python
for team in filt_roster:
  imp_team_df = filt_roster[team][0]
  opt_team_df = filt_roster[team][1]
  imp_team_df = imp_team_df[imp_team_df['Player'] != 'Team Total']
  opt_team_df = opt_team_df[opt_team_df['Player'] != 'Team Total']



  filt_roster[team] = imp_team_df, opt_team_df
```

```python
player_pos = {'Tremon Smith': 'CB', 'Quincy Williams': 'LB', 'Vincent Brown': 'WR', 'Sam Beal': 'CB', 'Tae Crowder': 'LB', 'Tytus Howard': 'OT', 'Jalen Thompson': 'DL', 'Michael Davis': 'CB', 'Virgil Green': 'TE', 'Adonis Alexander': 'CB'}
for team in filt_roster:

    # Identify players with null positions
    null_positions = filt_roster[team][1][filt_roster[team][1]['Pos'].isnull()]['Player']
    null_positions = filt_roster[team][0][filt_roster[team][0]['Pos'].isnull()]['Player']

    # Update positions based on the player_pos dictionary
    filt_roster[team][1]['Pos'] = filt_roster[team][1]['Pos'].fillna(null_positions.map(player_pos))
```
