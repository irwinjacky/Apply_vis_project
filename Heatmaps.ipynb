```python
import re

def extract_year(header):
    match = re.search(r'\b(\d{4})\b', header)
    if match:
        return match.group(1)
    else:
        return None

# Extracting the year from each column header
years = [extract_year(column) for column in inj_df.columns]
set_list = set(years)
set_list


move_col = inj_df.pop('Team')
inj_df.insert(1, 'Team', move_col)
inj_df
```

```python
player_df = pd.read_csv('/content/drive/MyDrive/Data_vis_proj/Important_Feats/important_roster_allteams.csv',index_col=False)
player_df

pos_df = player_df[['Player','Pos','Team']].copy()
pos_df.drop_duplicates()
```

```python
pos_df['Player'].value_counts()
```

```python
master_df = inj_df.merge(pos_df, left_on=['Player', 'Team'], right_on=['Player','Team'], how='inner')
master_df
```

```python
null_pos = master_df.loc[master_df['Pos'].isnull()]
null_pos

fill_pos = {'Vincent Brown': 'WR', 'Michael Davis': 'CB', 'Virgil Green':'TE', 'Sam Beal':'CB', 'Tae Crowder':'LB', 'Jalen Thompson':'S', 'Quincy Williams':'LB', 'Tytus Howard':'T'}

master_df['Pos'] = master_df['Pos'].fillna(master_df['Player'].map(fill_pos))
```
