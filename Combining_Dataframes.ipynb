```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os
```

```python
def read_multiple_csv_from_folder(folder_path, index_col=False):
    # Creating empty list to store dataframes.
    dfs = []

    # Checking if the folder exists.
    if os.path.exists(folder_path) and os.path.isdir(folder_path):
        # List all files in the folder with a '.csv' extension.
        csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

        for file in csv_files:
            file_path = os.path.join(folder_path, file)
            # Reading the CSV file into a DataFrame.
            df = pd.read_csv(file_path, index_col=False)
            # Removing first column because it's the original csv files index and is not needed.
            df = df.iloc[:, 1:]
            # Appending the DataFrame to the list
            dfs.append(df)
    else:
        print(f"Folder not found: {folder_path}")

    return dfs
```

```python
folder_imp_odds = "/content/drive/MyDrive/Data_vis_proj/important/odds"
imp_odds = read_multiple_csv_from_folder(folder_imp_odds)

folder_imp_roster = "/content/drive/MyDrive/Data_vis_proj/important/roster"
imp_rost = read_multiple_csv_from_folder(folder_imp_roster)

folder_imp_scores = "/content/drive/MyDrive/Data_vis_proj/important/scores"
imp_score = read_multiple_csv_from_folder(folder_imp_scores)

folder_imp_stats = "/content/drive/MyDrive/Data_vis_proj/important/stats"
imp_stat = read_multiple_csv_from_folder(folder_imp_stats)
```

```python
imp_odds
imp_odds_df = pd.concat(imp_odds)
imp_odds_df.reset_index(drop=True, inplace=True)
print('Important Odds:')
display(imp_odds_df)

print()

imp_rost
imp_rost_df = pd.concat(imp_rost)
imp_rost_df.reset_index(drop=True, inplace=True)
print('Important Roster:')
display(imp_rost_df)

print()

imp_score
imp_score_df = pd.concat(imp_score)
imp_score_df.reset_index(drop=True, inplace=True)
print('Important Score:')
display(imp_score_df)

print()

imp_stat
imp_stat_df = pd.concat(imp_stat)
imp_stat_df.reset_index(drop=True, inplace=True)
print('Important Stat:')
display(imp_stat_df)
```
